```python
import flopy
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import colors
from skimage.draw import polygon
```

```python
path = 'choshui_original.bas'      #boundaryæª”
l=0
bound = []
layer = []
b = []

with open(path) as f:
    for line in f.readlines():
        if 'Layer' in line:
            l += 1
        
        elif l>0:
            s = line.split(' ')

            for j in s:
                if j != '' :
                    b.append(int(j)) 

            if len(b) >= 309:
                layer.append(b)
                b=[]

            if len(layer) >= 400:
                bound.append(layer)
                layer = []

            if len(bound) == 9:
                break

bound = np.array(bound)
print(bound.shape)
```

```python
cmap = colors.ListedColormap(['blue','white','grey'])    #-1, 0, 1 

plt.figure(figsize=(9,12))
plt.pcolor(bound[0][::-1], cmap=cmap, edgecolors='black')
plt.show()
```

```python
vertices = np.array([[170,200],[168,195],[168,190],[167,180],[165,172],[160,170],[155,168],[148,166],[141,168],[136,170],
                     [131,172],[125,175],[122,178],[120,185],[118,190],[115,195],[106,192],[102,190],[95,189],[88,188],[84,189],
                     [80,190],[80,270],[170,270]
                    ])
rr,cc = polygon(vertices[:,0], vertices[:,1], bound[0][::-1].shape)

wel_list = []

for i in range(len(rr)):
    if bound[1][399-rr[i]][cc[i]] == 1:
        bound[0][399-rr[i]][cc[i]] = 2
        wel_list.append([399-rr[i],cc[i]])
```
