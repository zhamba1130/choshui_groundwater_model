```python
import flopy
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import colors
from skimage.draw import polygon
```

```python
path = 'choshui.bas'      #boundaryæª”
l=0
bound = []
layer = []
b = []

with open(path) as f:
    for line in f.readlines():
        if 'Layer' in line:
            l += 1
        
        elif l>0:
            s = line.split(' ')

            for j in s:
                if j != '' :
                    b.append(int(j)) 

            if len(b) >= 309:
                layer.append(b)
                b=[]

            if len(layer) >= 400:
                bound.append(layer)
                layer  =[]

            if len(bound) == 9:
                break


bound = np.array(bound)
print(bound.shape)
```
